```python
import pandas as pd
import numpy as np
import requests
import json
```

```python
r = requests.get('https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv')
with open('tweet_img_p.tsv',mode = 'wb') as file:
    file.write(r.content)

img_predictions = pd.read_csv('tweet_img_p.tsv',sep="\t")
```

```python
df_list = []
with open('tweet-json.txt') as json_file:
    for line in json_file.readlines():
        jsonObject = json.loads(line.strip())
        tweet_id = jsonObject['id']
        retweet_count = jsonObject['retweet_count']
        favorite_count = jsonObject['favorite_count']
        df_list.append({'tweet_id': tweet_id,
                        'retweet_count': retweet_count,
                        'favorite_count': favorite_count
        })
tweet_extras = pd.DataFrame(df_list, columns = ['tweet_id','retweet_count','favorite_count'])
```

## Clean

```python
twitter_archive_clean = twitter_archive.copy()
```

### Tidiness Issues:

#### Define

- Retweeted tweets donot contain ratings and should be removed from our dataset, this will be done using pandas drop method for column axis after filtering the dataset to only contain data that is not retweeted.

#### Code

```python
twitter_archive_clean = twitter_archive_clean[twitter_archive_clean['retweeted_status_id'].isnull()]
twitter_archive_clean.drop(['retweeted_status_id','retweeted_status_user_id','retweeted_status_timestamp'],axis=1, inplace=True)
```

#### Test

```python
twitter_archive_clean.head()
```

#### Define

- The 3 dog stage columns can be condensed into one (pupper, puppo, doggo) -> dog stage which contains a string indicating which stage the dog is at and turn floofer into boolean column

#### Code

```python
dog_stages = []
for index, row in twitter_archive_clean.iterrows():
    stage = np.nan
    if row['doggo']!= "None":
        stage = row['doggo']
    elif row['pupper'] != "None":
        stage = row['pupper']
    elif row['puppo'] != "None":
        stage = row['puppo']
    dog_stages.append(stage)
twitter_archive_clean['dog_stages'] = dog_stages
twitter_archive_clean.drop(['pupper','doggo','puppo'], axis = 1, inplace = True)
```

```python
twitter_archive_clean['floofer'] = (twitter_archive_clean['floofer'] != "None")
```

#### Test

```python
twitter_archive_clean.query('floofer == True')
```
