# Web Crawling Sample
- This sample uses the following website for the data-gathering purposes: 'http://books.toscrape.com/index.html'
- This notebook has the input and output pre-run for ease of use

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


def books_spider(max_pages):
    df_list = []
    page = 1
    while page <= max_pages:
        url = 'http://books.toscrape.com/catalogue/page-' + str(page)+ '.html'
        source_code = requests.get(url)
        plain_text = source_code.text
        soup = BeautifulSoup(plain_text)
        for article in soup.find_all('article', class_='product_pod'):
            book_title = article.find('h3').find('a')['title']
            book_price = article.find('div',class_='product_price').find('p',class_='price_color').text[2:]
            df_list.append({'book_title': book_title,
                    'book_price': float(book_price)})
        page += 1       
    return df_list
    
```

```python
# Number of pages used for testing is 3 , any number of pages can be used
df_list = books_spider(3)
df = pd.DataFrame(df_list, columns = ['book_title', 'book_price'])
```
